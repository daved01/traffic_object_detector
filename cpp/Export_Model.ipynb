```python
import torch
import torchvision

# Thanks to the dicussion at https://discuss.pytorch.org/t/fasterrcnn-resnet50-jit-trace/62337/11

# Select a model from torchvision
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True, progress=True, num_classes=91, 
                                                             pretrained_backbone=True, trainable_backbone_layers=3,
                                                            min_size=360)
model.eval()

# Use torch.jit.trace to create a traced jit model
traced_script_module = torch.jit.script(model)

```

```python
# Save the model for export into C++

torch.jit.export_opnames(traced_script_module)
```

```python
traced_script_module.save("fasterrcnn50fpn.pt")
```
